```python
hists = []bins = numpy.linspace(0,numpy.max(numpy.array(activities).flatten()),num_of_bins+1)for i in xrange(0,num_neurons):    for j in xrange(i,num_neurons):        if i != j:            iv= numpy.array(activities)[:,i].flatten()            jv= numpy.array(activities)[:,j].flatten()            hists.append(numpy.histogram2d(iv,jv,bins,normed=True)[0])
```
