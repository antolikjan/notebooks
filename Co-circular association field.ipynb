```python
import numpy
import pylab

%matplotlib inline



def plot_or(x,y,phi):
    d=0.02
    dx = numpy.cos(phi)*d
    dy = numpy.sin(phi)*d   
    pylab.plot([x+dx,x-dx],[y+dy,y-dy],'k')

def p(psi):
    pylab.figure(figsize=(5,5))
    pylab.xlim(-1,1)
    pylab.ylim(-1,1)
    for i in xrange(0,30):
        for j in xrange (0,30):
            x = (i-15)/15.0
            y = (j-15)/15.0
            phi = numpy.arctan2(y,x)#
            orr = 2*phi-psi
            plot_or(x,y,orr)
        
    pylab.show()

#p(0)
p(numpy.pi/4)
#p(numpy.pi/2)
#p(numpy.pi/4*3)
#p(numpy.pi)
#p(numpy.pi/4*5)
#p(numpy.pi/4*6)
#p(numpy.pi/4*7)
#p(2*numpy.pi)
print numpy.arctan2(0,0)
```
